```python
import pandas as pd
import numpy as np
from sklearn import cluster
```

```python
stUrl = "https://raw.githubusercontent.com/hyokwan/python-lecture/master/dataset/student_middle.csv"
studentData = pd.read_csv(stUrl)
```

### 1. 타입통합 및 숫자컬럼 추가

```python
# studentData에서 숫자형 컬럼만 갖고 오기.
# studentData.select_dtypes(np.number).columns
# 해당 값을 list에 담기
features = list(studentData.select_dtypes(np.number).columns)
```

```python
features = list( studentData.select_dtypes(np.number).columns )
featureData = studentData.loc[ :, features ]
```

### 2. 모델적용

```python
from sklearn.cluster import KMeans
```

```python
### 모델 선언
kmeansModel = KMeans( random_state = 10, n_clusters=2)
```

```python
### 자율 학습
fittedKmeans = kmeansModel.fit(X=featureData)
```

```python
kmeansModel = cluster.KMeans( n_clusters=2, random_state=1 )
kmeansModel
```

```python
fittedKmeansModel = kmeansModel.fit(featureData)
```

### 3. 자율학습 예측

```python
predictValue = fittedKmeansModel.predict( featureData )
predictValue
```

```python
studentData["KMEANS_PREDICT"] = predictValue
```

### 4. 데이터 정리

```python
studentData["KMEANS_PREDICT"] = predictValue
studentData
```
