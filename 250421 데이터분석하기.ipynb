### 2. 특성선정 및 데이터 분리

```python
# 상관계수 기반 특성 선택
### 특성선정
corrDf = featuresDf.corr()
label = ["sales"]
stdCorr = 0.3
features = list (corrDf.loc[(abs(corrDf.sales) >= stdCorr) & (abs(corrDf.sales) != 1)].index )
```

```python
print( f'선정된 features: {features}')
print( f'선정된 label: {label}')
```

```python
### 데이터 분리
from sklearn.model_selection import train_test_split

trainData, testData = train_test_split( featuresDf, test_size = 0.2, random_state = 10)
```

```python
from sklearn.model_selection import train_test_split
trainData, testData = \
    train_test_split( featuresDf, test_size=0.2, random_state=10)
trainDataFeatures = trainData.loc[:, features]
trainDataLabel = trainData.loc[:, label]
trainDataAll = trainData.loc[:]
testDataFeatures = testData.loc[:, features]
testDataLabel = testData.loc[:, label]
testDataAll = testData.loc[:]
print( trainDataFeatures.shape )
print( trainDataLabel.shape )
print( testDataFeatures.shape )
print( testDataLabel.shape )
print( trainDataAll.shape )
print( testDataAll.shape )
```

### 3. 모델 적용

```python
# 필요한 라이브러리 임포트
from sklearn.tree import DecisionTreeRegressor        # 결정트리 회귀 모델
from sklearn.ensemble import RandomForestRegressor    # 랜덤포레스트 회귀 모델
from sklearn.linear_model import LinearRegression     # 선형회귀 모델
from sklearn.tree import plot_tree                    # 트리 시각화 함수
import matplotlib.pyplot as plt                       # 시각화를 위한 matplotlib
```

```python
### 모델 정의
```

```python
# 결정 트리 회귀 모델 생성 (max_depth=10으로 과적합 방지, random_state는 동일한 결과를 위해 고정)
dtModel = DecisionTreeRegressor(random_state=10, max_depth=10)

# 랜덤 포레스트 회귀 모델 생성 (20개의 결정 트리를 사용)
rfModel = RandomForestRegressor(random_state=10, n_estimators=20)

# 선형 회귀 모델 생성
lrModel = LinearRegression()
```

```python
# 2. 데이터 조건 필터 예시 (debt가 10118 이하인 데이터의 개수 확인)
# 특정 조건의 데이터를 필터링하여 확인할 때 사용
trainDataFeatures.loc[trainDataFeatures.debt <= 10118].shape
```

```python
# 결정 트리 모델 학습
fittedDtModel = dtModel.fit(X=trainDataFeatures, y=trainDataLabel)

# 랜덤 포레스트 모델 학습
fittedRfModel = rfModel.fit(X=trainDataFeatures, y=trainDataLabel)

# 선형 회귀 모델 학습
fittedLrModel = lrModel.fit(X=trainDataFeatures, y=trainDataLabel)
```

```python
# 시각화
# 큰 그림을 위해 figsize 설정 (폭이 250인 긴 가로 시각화)
plt.figure(figsize=(250, 15))

# 학습된 결정 트리 모델을 시각화
plot_tree(fittedDtModel, feature_names=trainDataFeatures.columns, filled=True)

# 그림을 파일로 저장 (너무 커서 화면에 바로 띄우지 않고 저장)
plt.savefig("./plt_dt.png")
```
