```python
import pandas as pd
import matplotlib.pyplot as plt

featuresLocalUrl = "https://raw.githubusercontent.com/hyokwan/python-lecture/master/dataset/feature_regression_example.csv"
featuresData = pd.read_csv(featuresLocalUrl)

featuredData2015 = featuresData.loc[ featuresData.YEAR == 2015 ]

x = featuredData2015["WEEK"]
y = featuredData2015["QTY"]
y2 = featuredData2015["PRO_PERCENT"]

fig=plt.figure(figsize=(14,4))
ax1 = fig.add_subplot(1,1,1)
ax1.plot(x,y, "bo-", label="QTY")
ax1.set_xlabel("week")
ax1.set_ylabel("qty")
ax1.legend(loc=2)

ax2 = ax1.twinx()
ax2.plot(x,y2, "ro-", label="PRO_PERCENT")
ax2.set_ylabel("pro_percent")
ax2.legend(loc=1)
```
