```python
# git 불러오기
import pandas as pd
url = 'https://raw.githubusercontent.com/hyokwan/python-lecture/refs/heads/master/dataset/feature_regression_example.csv'
df = pd.read_csv(url)

groupKey = [ 'REGIONID', 'PRODUCT', 'WEEK']
groupData = df.groupby(by=groupKey)['QTY'].agg(['mean']).reset_index()
groupData = groupData.rename(columns = {"mean":"QTY_MEAN"})

import matplotlib.pyplot as plt
import seaborn as sns

fig = plt.figure( figsize=(18,6) )
ax1 = fig.add_subplot(1, 2, 1)
sns.histplot(data=groupData, x=groupData.QTY_MEAN, ax=ax1)
ax2 = fig.add_subplot(1, 2, 2)
sns.kdeplot(data=groupData, x=groupData.QTY_MEAN)
```
