```python
# git 불러오기
import pandas as pd
url = 'https://raw.githubusercontent.com/hyokwan/python-lecture/refs/heads/master/dataset/feature_regression_example.csv'
df = pd.read_csv(url)
```

```python
# 2015년도 데이터 분리하기
target = df.loc[df.YEAR == 2015]

```

```python
#필요 라이브러리 정의
import matplotlib.pyplot as plt

#차트 정의하기
x = target['WEEK']
y = target['QTY']

plt.plot('2015년 QTY')
plt.plot(x, y, 'bo-', label='QTY', lw=1.5)
plt.legend(loc=0)
```
